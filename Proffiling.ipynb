```python
import pandas as pd
```

```python

# Load the dataset
data = pd.read_csv(r'final_df.csv')
```

```python
data.month.unique()
```

```python
data.year.unique()
```

```python
# Group the observations by month and year
grouped = data.groupby(['month', 'year'])

```

```python
len(grouped)
```

```python


# Create a new DataFrame with the desired columns and rows
profiled_data = pd.DataFrame({
    'min_Predicted_TOTAL': grouped['Predicted_TOTAL'].min(),
    'avg_Predicted_TOTAL': grouped['Predicted_TOTAL'].mean(),
    'max_Predicted_TOTAL': grouped['Predicted_TOTAL'].max(),
    'min_WIND TOTAL': grouped['WIND TOTAL'].min(),
    'avg_WIND TOTAL': grouped['WIND TOTAL'].mean(),
    'max_WIND TOTAL': grouped['WIND TOTAL'].max(),
    'min_GEOTHERMAL': grouped['GEOTHERMAL'].min(),
    'avg_GEOTHERMAL': grouped['GEOTHERMAL'].mean(),
    'max_GEOTHERMAL': grouped['GEOTHERMAL'].max(),
    'min_total_renewable': grouped['total_renewable'].min(),
    'avg_total_renewable': grouped['total_renewable'].mean(),
    'max_total_renewable': grouped['total_renewable'].max(),
    'min_SMALL HYDRO': grouped['SMALL HYDRO'].min(),
    'avg_SMALL HYDRO': grouped['SMALL HYDRO'].mean(),
    'max_SMALL HYDRO': grouped['SMALL HYDRO'].max(),
    'min_BIOMASS': grouped['BIOMASS'].min(),
    'avg_BIOMASS': grouped['BIOMASS'].mean(),
    'max_BIOMASS': grouped['BIOMASS'].max()
},  index=grouped.indices.keys())
```

```python
profiled_data.head(46)
```

```python
profiled_data.tail(48)
```

```python
# Drop rows where 'min_Predicted_TOTAL' is negative

profiled_data_filtered = profiled_data[profiled_data['min_Predicted_TOTAL'] >= 0]
```

```python
profiled_data_filtered
```

```python
profiled_data_filtered.to_csv('profilled_df.csv', index = True)
```
